```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib as mpl 
import numpy as np
from scipy import stats 
from scipy.stats import norm, skew
```

```python
data = pd.read_csv("/home/andrew/ML/Tutorials/data/WA_Fn-UseC_-HR-Employee-Attrition.csv")
data.head().transpose()
```

```python
data.shape
```

```python
# exploratory data analysis: an approach of analyzing dataset to summarize their main characteristics. normall 
# using visual methods 
data.columns
```

```python
# first we pick our target variable which in turn inform us on which model to use:
# to remove features that will not necessarily increase model accuracy
del_feutures = ['EmployeeNumber','Gender','HourlyRate','DailyRate','MaritalStatus','OverTime','RelationshipSatisfaction',
               'TrainingTimesLastYear','WorkLifeBalance','YearsWithCurrManager','Over18','EmployeeCount','MonthlyRate','StandardHours']
df1 = data.drop(labels=del_feutures,axis=1)
df1.shape
```

```python
mapping = {'Yes':1,'No':0}
df1.Attrition = df1.Attrition.apply(lambda x: mapping[x])
```

```python
#visualizations : using seaborn 
categoric_cols = df1.select_dtypes(np.object)
categoric_cols.columns
```

```python
for col in categoric_cols:
    fig = plt.figure(figsize=(8,6))# defines the plot area 
    sns.countplot(x=categoric_cols[col],data=df1,hue='Attrition')
    plt.xticks(rotation ='vertical')
    plt.title("BAR PLOT FOR ATTRITION AGANAIST: {}".format(col))
```

```python
# modelling  data
#first we transform categoric to numeric data 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelEncode = LabelEncoder()
for i in categoric_cols:
    categoric_cols[i] = labelEncode.fit_transform(categoric_cols[i])
categoric_cols.head()
```

```python
# we concat datasets for both numeric and categoric_transformed 
final_data = pd.concat([numeric_cols,categoric_cols],axis=1)
final_data.shape
```

```python
final_data.columns
```

```python
# we divide data into training and test split 
from sklearn.model_selection import train_test_split
#we separate our target variable from the dataset 
y = final_data.pop('Attrition')
X_train, X_test, Y_train, Y_test = train_test_split(final_data,y,test_size =0.2)
```

```python
#kNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,Y_train)
knn_score_train = knn.score(X_train,Y_train)
knn_score_test = knn.score(X_test,Y_test)
print("KNN Train Score: {} KNN Test Score: {}".format(knn_score_train,knn_score_test))
```

```python
# Logistic Regression 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
lr_score_train = lr.score(X_train,Y_train)
lr_score_test = lr.score(X_test,Y_test)
print("Logistic Regression Train Score: {} Test Score: {}".format(lr_score_train,lr_score_test))
```

```python
#SVM
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, Y_train)
svm_score_train = svm.score(X_train, Y_train)
print("Training score: ",svm_score_train)
svm_score_test = svm.score(X_test, Y_test)
print("Testing score: ",svm_score_test)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
rfc_score_train = rfc.score(X_train, Y_train)
rfc_score_test = rfc.score(X_test, Y_test)
print("Training Score : {} Testing Score: {} ".format(rfc_score_train,rfc_score_test))
```

```python
# feuture imporntance according to Random Forest 
feat_importances = pd.Series(rfc.feature_importances_, index=final_data.columns)
feat_importances = feat_importances.nlargest(15)
fig = plt.figure(figsize=(12,10))
feat_importances.plot.bar()
```
